```
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```
kcp2 = pd.read_csv('/content/kcp_coords2.csv')
kcp2.head()
```

```
kcp_coord = [43.7285,-79.4322 ]
my_map = folium.Map(location = kcp_coord, zoom_start = 15)
for i in range(len(kcp2)):
  #print(kcp2['lat'][i], kcp2['long'][i], kcp2['Street Address'][i])
  folium.Marker([kcp2['lat'][i], kcp2['long'][i]], popup = kcp2['Street Address'][i]).add_to(my_map)

my_map
```
